```python
from gramformer import Gramformer
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('new_review.csv')
df
```

```python
df=df.drop(['star','app_id','reviewDate'],axis=1)
df
```

```python
gf=Gramformer(models=3)
```

```python
def set_seed(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

set_seed(1212)


gf = Gramformer(models = 1, use_gpu=True) # 1=corrector, 2=detector

influent_sentences = df   
l=[]
k=[]
for influent_sentence in df["text"]:
    corrected_sentences = gf.correct(influent_sentence, max_candidates=1)
    #print('[Input] ',influent_sentence)
    #output=''
    for corrected_sentence in corrected_sentences:
      #print('[Correction] ',corrected_sentence)
      if (influent_sentence != corrected_sentence):
        l.append(corrected_sentence)
        k.append(influent_sentence)
        #print('[Correction] ',corrected_sentence)
        #print('Correct one')
    #print("-" *100)
#data=[influent_sentence] + [corrected_sentences]
```
